```python
# constructing all 64 basis elements
basis = []

# the below j is the imaginary i

j = complex(0,1)

# below function allows us to do tensor product of three operators
def triple(A,B,C):
    temp = np.kron(A,B)
    temp = np.kron(temp,C)
    return temp

#pauli matrices
I = np.array([[1,0],[0,1]])
sx = np.array([[0,1],[1,0]])
sy = np.array([[0,-j],[j,0]])
sz = np.array([[1,0],[0,-1]])

#list of pauli matrices below
p = [I,sx,sy,sz]

counti = 0
while counti < 4:
    countj = 0
    while countj < 4:
        countk = 0
        while countk < 4:
            basis.append(triple(p[counti],p[countj],p[countk])/(math.sqrt(8)))
            countk = countk + 1
        countj = countj + 1
    counti = counti + 1
    
```

```python
# matrix dot product

def matrix_dot(a,b):
    
    return np.trace(np.dot(a,b))
```
